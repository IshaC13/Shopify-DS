# Question 1:
On Shopify, we have exactly 100 sneaker shops, and each of these shops sells only one model of shoe. We want to do some analysis of the average order value (AOV). When we look at orders data over a 30 day window, we naively calculate an AOV of $3145.13. Given that we know these shops are selling sneakers, a relatively affordable item, something seems wrong with our analysis.

## Question 1.a :
Think about what could be going wrong with our calculation. Think about a better way to evaluate this data. 


```
data.describe()
```

1. The average order value(AOV) provided in the question is $3145.13 which is obviously pricy for sneakers. So after checking the order_amount we see that the amount was obtained by taking the average of the order_amount column. Thus taking the mean is not a very good estimate of the AOV.
2. The median here is 284 which is much more reasonable and the actual mean should lie between those two.

After looking at the 3rd quartile, 75% of the data points fall under 390. This amount is very less when compared to the AOV above. While the max value is 704000 which is way larger  than the 3rd quartile. So we can conclude here that outliers contribute a lot in the AOV calculation.

So instead of AOV we can use Median Order Value(MOV). While performing data analysis on most of the dataset such as this it is good to check the MOV and use it as the 'average' function when there are outliers present in the data. The large and small outliers do not skew the meadian as much as mean so it is better to use MOV.

```
# Displaying Average and Median Order Value
print('Average Order Value:', data['order_amount'].mean(), ',',
      'Median Order Value:', data['order_amount'].median())
```

## Question 2.b :
What is the last name of the employee with the most orders?


### Answer to 2.b:

Query used for the question:- 

**SELECT LastName FROM Employees
WHERE EmployeeID = (SELECT TOP 1 EmployeeID FROM
(SELECT EmployeeID, COUNT(OrderID) FROM Orders GROUP BY EmployeeID ORDER BY COUNT(OrderID) DESC));**

Answer:- **Peacock**



## Question 2.c :
What product was ordered the most by customers in Germany?


### Answer to 2.c:

Query used for the question:- 

**SELECT ProductName FROM Products
WHERE ProductID IN (SELECT TOP 1 ProductID FROM OrderDetails
WHERE OrderID IN (
SELECT OrderID FROM Orders
WHERE CustomerID IN (SELECT CustomerID FROM Customers
WHERE Country = 'Germany'))
GROUP BY ProductID
ORDER BY COUNT(OrderDetailID) DESC);**
